# Replace outliers in numeric columns of a dataframe.

```python
def replace_outliers(df, numeric_cols, method="median"):
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if method == "median":
            median = df[col].median()
            df[col] = df[col].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
        elif method == "cap":
            df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
        else:
            raise ValueError("Method must be 'median' or 'cap'")
    return df
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()

# Replace outliers with median
df_cleaned = replace_outliers(df, numeric_cols, method="median")
```

```python
# Step 1: Standardization
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
# Drop non-numeric columns like 'Country'
X = df.drop(columns=["Country"])
```

```python
# Handle missing values (replace with mean)
imputer = SimpleImputer(strategy="mean")
X_imputed = imputer.fit_transform(X)
```

```python
# Scale features (important for clustering so all variables are comparable)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
print("Data prepared for clustering. Shape:", X_scaled.shape)
```

```python
# KMeans Clustering
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Elbow method to find best k
sse = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    sse.append(km.inertia_)
```

```python
plt.plot(range(2, 11), sse, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("SSE")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
# Reduce data to 2 dimensions using PCA for visualization
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Final KMeans model with k=3 
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
print("KMeans Clusters:", set(kmeans_labels))
```

```python
from sklearn.metrics import silhouette_score

# KMeans with k=3 (already fitted)
kmeans_labels = kmeans.labels_
kmeans_score = silhouette_score(X_scaled, kmeans_labels)
print("Silhouette Score (KMeans, k=3):", kmeans_score)
```

```python
# Hierarchical Clustering
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
# Perform hierarchical clustering
linked = linkage(X_scaled, method='ward')
```

```python
# Dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, truncate_mode="lastp", p=20, show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Agglomerative Clustering with 3 clusters
agg = AgglomerativeClustering(n_clusters=3, linkage="ward")
agg_labels = agg.fit_predict(X_scaled)

# Silhouette score
agg_score = silhouette_score(X_scaled, agg_labels)
print("Silhouette Score (Hierarchical, k=3):", agg_score)
```

```python
# DBSCAN Clustering
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
# -1 means noise/outliers
print("DBSCAN Clusters:", set(dbscan_labels))  
```

```python
# Evaluate performance
if len(set(dbscan_labels)) > 1:
    score = silhouette_score(X_scaled, dbscan_labels)
    print("Silhouette Score (DBSCAN):", score)
else:
    print("Silhouette Score (DBSCAN): Not valid (only 1 cluster)")
```

```python
best_score = -1
best_params = None
```

```python
# Try different eps (neighborhood radius) and min_samples
for eps in [0.3, 0.5, 0.75, 1.0, 1.5]:
    for min_samples in [3, 5, 10]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)

        # Check valid clustering (at least 2 clusters, not only noise)
        if len(set(labels)) > 1 and len(set(labels)) != 1:
            try:
                score = silhouette_score(X_scaled, labels)
                print(f"eps={eps}, min_samples={min_samples}, score={score:.4f}")
                if score > best_score:
                    best_score = score
                    best_params = (eps, min_samples)
            except:
                pass
```

```python
print("\nBest DBSCAN Silhouette Score:", best_score, "with params:", best_params)
```
